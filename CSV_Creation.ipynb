```python
import pandas as pd

# Load the dictionary from the file
with open('/content/Proff_Info.txt', 'r') as f:
    professors_dict = eval(f.read())

# Create an empty list to store rows
rows = []

# Iterate over each department and professor in the dictionary
for department, professors in professors_dict.items():
    for professor, research_areas in professors.items():
        # If a professor has multiple research areas, split their information into multiple rows
        if len(research_areas) > 1:
            for research_area in research_areas:
                rows.append([professor, department, research_area])
        else:
            rows.append([professor, department, research_areas[0]])

# Create a pandas DataFrame from the list of rows
df = pd.DataFrame(rows, columns=['Name of the Professor', 'Department', 'Research Area'])

# Save the DataFrame as a CSV file
df.to_csv('professors.csv', index=False)
```

```python
def is_substring_present(substring, string_list):
    for string in string_list:
        if substring in string:
            return True
    return False
```

```python
import pandas as pd
import pickle

# Load the "B" dictionary from the pickle file
with open('/content/Research_Domain_Research_Area_Dictionary.pickle', 'rb') as f:
    b_dict = pickle.load(f)

# Load the "professors.csv" file into a pandas DataFrame
df = pd.read_csv('professors.csv')

# Create a new "Broad Research Area" column in the DataFrame
df['Broad Research Area'] = ''

# Iterate over each row in the DataFrame
for i, row in df.iterrows():
    research_area = row['Research Area']
    broad_area = ''
    # Use the "B" dictionary to find the broad topic for each subtopic
    if research_area=='Coming soon...':
      broad_area = 'Not Available'
      df.at[i, 'Broad Research Area'] = broad_area
      continue

    for broad, subtopics in b_dict.items():
        if is_substring_present(research_area.lower(),subtopics):
            broad_area = broad
            break
    # Update the "Broad Research Area" column with the broad topic for each subtopic
    df.at[i, 'Broad Research Area'] = broad_area

# Drop the "Research Area" column
df = df.drop(columns=['Research Area'])

# Drop the duplicate rows in the DataFrame
df = df.drop_duplicates()

# Save the final DataFrame as a CSV file
df.to_csv('final_professors2.csv', index=False)
```

```python
import pandas as pd

# Load the "final_professors2.csv" file into a pandas DataFrame
df = pd.read_csv('final_professors2.csv')

# Drop the rows where "Broad Research Area" column is empty
df = df.dropna(subset=['Broad Research Area'])

# Save the updated DataFrame as a CSV file
df.to_csv('updated_professors.csv', index=False)
```
